```python
cfram=df1[df1['a'].notnull()]
type(cfram)
```

```python
#cfram1=df1[df1.a.notnull()]
#type(cfram1)
```

```python
operating_system1=cfram.a.str.contains('Windows')  # contins marks records as true or false if condition matches

print(type(operating_system1))
operating_system1
#np.where(operating_system,'Windows', 'Not Windows')
```

```python
operating_system=np.where(operating_system1,'Windows', 'Not Windows')
```

```python
print(type(operating_system))
operating_system[:10]
```

```python
by_tz_os=cfram.groupby(['tz',operating_system])
by_tz_os
```

```python
by_tz_os.size().unstack().fillna(0)
```

```python
agg_counts=by_tz_os.size().unstack().fillna(0)
type(agg_counts)
```

```python
index=agg_counts.sum()
index
```

```python
#by_tz_os.size().fillna('0')
```

```python
indexer=agg_counts.sum(1).argsort()
indexer
```

```python
indexer[:10]
```

```python
count_subset=agg_counts.take(indexer)[-10:]
count_subset
```

```python
count_subset.plot(kind='barh',stacked='true')
```
